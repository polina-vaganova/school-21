### make the aggregations with the `Make` and the `Model`:

* display the median `Fines` grouped by the `Make`:

``````python
df.groupby("Make", )["Fines"].median().reset_index().rename(columns={"Fines": "FinesMedian", }, )
``````

* display the median `Fines` grouped by the `Make` and the `Model`:

``````python
df.groupby(["Make", "Model", ], )["Fines"].median().reset_index().rename(columns={"Fines": "FinesMedian", }, )
``````

* display the number of `Fines` grouped by the `Make` and the `Model` in order to understand if we can trust the median values:

``````python
df.groupby(["Make", "Model", ], )["Fines"].count().reset_index().sort_values("Fines", ).rename(columns={"Fines": "FinesNumber", }, )
``````

* display the minimum and the maximum `Fines` grouped by the `Make` and the `Model` in order to better understand the variance:

``````python
df.groupby(["Make", "Model", ], )["Fines"].min().reset_index().sort_values("Fines", ).rename(columns={"Fines": "FinesMin", }, )
``````

``````python
df.groupby(["Make", "Model", ], )["Fines"].max().reset_index().sort_values("Fines", ).rename(columns={"Fines": "FinesMax", }, )
``````

* display the standard deviation of the `Fines` grouped by the `Make` and the `Model` in order to better understand the variance:

``````python
df.groupby(["Make", "Model", ], )["Fines"].std().reset_index().sort_values("Fines", ).rename(columns={"Fines": "FinesSTD", }, )
``````

### make the aggregations with the `CarNumber`:

* display the `CarNumber` grouped by the number of the `Fines` in descending order, we want to find those who most often violated the law:

``````python
df.groupby(df.index, )["Fines"].count().reset_index().sort_values(by=["Fines", ], ascending=False, ).set_index("CarNumber", drop=True, ).rename(columns={"Fines": "FinesNumber", }, )
``````

* select from the initial dataframe all the rows corresponding to the top `1` `CarNumber`, we want to zoom in a little bit:

Select top `1` `CarNumber`:

``````python
top_car_num: str = df.groupby(df.index, )["Fines"].count().reset_index().sort_values(by=["Fines", ], ascending=False, ).set_index("CarNumber", drop=True, ).iloc[0].name
``````

Select suitable rows:

``````python
df[df.index == top_car_num]
``````

* display the `CarNumber` grouped by the sum of the `Fines` in descending order, we want to find those who paid the most:

``````python
df.groupby(df.index, )["Fines"].sum().reset_index().sort_values(by=["Fines", ], ascending=False, ).set_index("CarNumber", drop=True, ).rename(columns={"Fines": "FinesSum", }, )
``````

* select from the initial dataframe all the rows corresponding to the top `1` `CarNumber`, we want to zoom in a little bit:

Select top `1` `CarNumber`:

``````python
top_car_num: str = df.groupby(df.index, )["Fines"].sum().reset_index().sort_values(by=["Fines", ], ascending=False, ).set_index("CarNumber", drop=True, ).iloc[0].name
``````

Select suitable rows:

``````python
df[df.index == top_car_num]
``````

Calculate count of unique car models for every `CarNumber`:
